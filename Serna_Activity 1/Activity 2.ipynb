<h3>FILLING UP NULL VALUES</h3>

```python
accidents_df.isnull().sum().reset_index()
```

```python
accidents_df['Latitude'].mean()
accidents_df['Latitude'] =accidents_df['Latitude'].fillna(accidents_df['Latitude'].mean())
accidents_df['Longitude'].mean()
accidents_df['Longitude'] = accidents_df['Longitude'].fillna(accidents_df['Longitude'].mean())
```

```python
accidents_df['Road_Surface_Conditions'].mode()
accidents_df['Road_Surface_Conditions'] = accidents_df['Road_Surface_Conditions'].fillna(accidents_df['Road_Surface_Conditions'].mode()[0])
```

```python
accidents_df['Road_Type'].mode
accidents_df['Road_Type'] = accidents_df['Road_Type'].fillna(accidents_df['Road_Type'].mode()[0])
```

```python
accidents_df['Urban_or_Rural_Area'].mode()
accidents_df['Urban_or_Rural_Area'] = accidents_df['Urban_or_Rural_Area'].fillna(accidents_df['Urban_or_Rural_Area'].mode()[0])
```

```python
accidents_df['Weather_Conditions'].mode()
accidents_df['Weather_Conditions'] = accidents_df['Weather_Conditions'].fillna(accidents_df['Weather_Conditions'].mode()[0])
```

```python
accidents_df.isnull().sum()
```

```python
accidents_df.info()
```

```python
accidents_df.columns
```

```python
accidents_df.dtypes.reset_index()
```

```python
accidents_df['Accident Date'] = pd.to_datetime(accidents_df['Accident Date'], dayfirst = True, errors ='coerce')
accidents_df['Year'] = accidents_df['Accident Date'].dt.year
accidents_df['Month'] = accidents_df ['Accident Date'].dt.month
accidents_df['Day'] = accidents_df ['Accident Date'].dt.day
accidents_df['DayOfWeek'] = accidents_df ['Accident Date'].dt.dayofweek
```

```python
accidents_df.dtypes.reset_index()
```

```python
accidents_df['Index'] = accidents_df['Index'].astype('category')
accidents_df['Accident_Severity'] = accidents_df['Accident_Severity'].astype('category')
accidents_df['Light_Conditions'] = accidents_df ['Light_Conditions'].astype('category')
accidents_df['District Area'] = accidents_df ['District Area'].astype('category')
accidents_df['Road_Surface_Conditions'] = accidents_df ['Road_Surface_Conditions'].astype('category')
accidents_df['Road_Type'] = accidents_df['Road_Type'].astype('category')
accidents_df['Urban_or_Rural_Area'] = accidents_df['Urban_or_Rural_Area'].astype('category')
accidents_df['Weather_Conditions'] = accidents_df['Weather_Conditions'].astype('category')
accidents_df['Vehicle_Type'] = accidents_df['Vehicle_Type'].astype('category')
```

<h3>INSIGHTS #29: District Clackmannanshire,Orkney Islands,Shetland Islands, Clackmannshire and London Airport(Heathrow) are the five districts that has low numbers of casualties. District Clackmannanshire has the lowest casualties recorded.</h3>

<h3>Number of casualties of the 5 district that has the lowest caccidents recorded </h3>

```python
districts = ["Clackmannanshire", "Orkney Islands", "Shetland Islands", "London Airport (Heathrow)"]
severity_counts = (accidents_df[accidents_df["District Area"].isin(districts)].groupby(["District Area", "Accident_Severity"]).size().reset_index(name="Count"))
plt.figure(figsize=(10,6))
for district in districts:
    subset = severity_counts[severity_counts["District Area"] == district]
    labels = subset["Accident_Severity"].astype(str) + " (" + district + ")"
    plt.bar(labels, subset["Count"], label=district)
    plt.title("Accident Severity in Selected Districts")
    plt.ylabel("Number of Accidents")
    plt.xticks(rotation=45, ha="right")
    plt.legend(title="Districts")
    plt.tight_layout()
    plt.show()
```

<h3>INSIGHTS #30:In Clackmannanshire, Orkney Islands, Shetland Islands, and London Airport (Heathrow), most accidents recorded were slight, showing they were generally less severe. Heathrow had the highest overall number of slight cases (168), while Clackmannanshire followed with 72. Serious accidents were relatively low across all districts, ranging from 9 in Heathrow to 19 in Orkney. Fatal cases were rare, but Shetland Islands stood out with 5 fatalities, the highest among the four areas, while the others recorded only 1–2 each. This suggests that while accidents are common, especially slight ones, Shetland has a comparatively higher share of fatal outcomes.</h3>
