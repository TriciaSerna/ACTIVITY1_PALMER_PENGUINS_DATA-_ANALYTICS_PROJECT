```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
crime_df['	Location Description'] = crime_df['Location'].fillna('Unknown')
crime_df['Location'] = crime_df['Location'].fillna('Unknown')
```

```python
crime_df['Community Area'].mean()
crime_df['Community Area'] =crime_df['Community Area'].fillna(crime_df['Latitude'].mean())
crime_df['X Coordinate'].mean()
crime_df['X Coordinate'] =crime_df['X Coordinate'].fillna(crime_df['X Coordinate'].mean())
crime_df['Y Coordinate'].mean()
crime_df['Y Coordinate'] =crime_df['Y Coordinate'].fillna(crime_df['Y Coordinate'].mean())
crime_df['Latitude'].mean()
crime_df['Latitude'] =crime_df['Latitude'].fillna(crime_df['Latitude'].mean())
crime_df['Longitude'].mean()
crime_df['Longitude'] =crime_df['Longitude'].fillna(crime_df['Longitude'].mean())
```

```python
crime_df.isnull().sum()
```

```python
crime_df.info()
```

```python
crime_df.dtypes.reset_index()
```

```python
crime_df['Date'] = pd.to_datetime(crime_df['Date'], dayfirst = True, errors ='coerce')
crime_df['Year'] = crime_df['Date'].dt.year
crime_df['Month'] = crime_df ['Date'].dt.month
crime_df['Day'] = crime_df ['Date'].dt.day
crime_df['DayOfWeek'] = crime_df ['Date'].dt.dayofweek
```

```python
crime_df['Case Number'] = crime_df['Case Number'].astype('category')
```

```python
crime_df['Block'] = crime_df['Block'].astype('category')
crime_df['IUCR'] = crime_df['IUCR'].astype('category')
crime_df['Primary Type'] = crime_df['Primary Type'].astype('category')
crime_df['Description'] = crime_df['Description'].astype('category')
crime_df['Location Description'] = crime_df['Location Description'].astype('category')
crime_df['FBI Code'] = crime_df['FBI Code'].astype('category')
crime_df['Updated On'] = crime_df['Updated On'].astype('category')
crime_df['Location'] = crime_df['Location'].astype('category')
crime_df['\tLocation Description'] = crime_df['\tLocation Description'].astype('category')
```

<h3>Q32. In 2024,  which location description in Area 25 had the worst arrest-to-crime ratio?</h3>

```python
df_top = crime_df[(crime_df['Community Area']==top_area) & (crime_df['Year']==top_year)]
loc_stats = df_top.groupby('Location Description').agg(total=('ID','count'), arrests=('Arrest','sum'))
loc_stats = loc_stats[loc_stats['total']>=10]  # filter small counts
loc_stats['arrest_rate'] = (loc_stats['arrests'] / loc_stats['total'])*100
loc_stats.sort_values('arrest_rate').head(10)['arrest_rate'].plot(kind='bar', figsize=(12,5))
plt.title(f"Lowest Arrest Rates by Location in {top_area} ({top_year})")
plt.ylabel("Arrest Rate (%)")
plt.show()
```

<h3> INSIGHTS:In 2024, Area 25 had  low arrest-to-crime ratios at several locations, with the worst rates seen at banks, driveways, and medical/dental officesâ€”each reporting arrest rates below 2%. This means that crimes occurring in these locations were less likely to result in arrests, potentially indicating enforcement or reporting challenges specific to those environments.</h3>
